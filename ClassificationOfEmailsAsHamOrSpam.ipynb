<h1>CLASSIFICATION OF EMAILS AS HAM OR SPAM<h1>

<h2>INTRODUCTION<h2>

>Email is an extremely common form of communication today, especially for official notifications and in the corporate setting. It is also frequently used by online sites to the point that several people have tens of thousands of unread emails. When companies want to promote new items or announce offers and sales, they send emails to consumers in bulk, and such emails are known as 'spam mails'.
>Spam mails are undesirable-they take up a very large amount of storage space, fill up the inbox, create unnecessary clutter, and significantly reudce the performance of the device. Spam mails may contain genuine offers from trusted sources, but due to the frequency, a spam mail containing malware can be easily misunderstood by the user for an actual offer, leading to the user's getting trapped with no way out.
>Cybercriminals employ spam mails to send malicious attachments or links to regular, unsuspecting consumers in the disguise of promotional mails. As soon as the attachment/link is clicked, the malware gets downloaded in the system, providing unauthorized access of the user's most personal details and activities to the attacker. The attacker can use this information to cause very serious harm to the user's reputation and permanently damage the user's financial record history, and can even make the user go bankrupt by threatening to leak information.
>For all of these reasons, it is absolutely a compulsion for all systems and devices to have a robust spam mail filter enabled. Such a filter should be able to detect spam mails with extremely high accuracy and immediately shift such an email to a separate folder to protect the user's data and privacy.
>In this project, a <b>MACHINE LEARNING MODEL</b> will be created using <code>Logistic Regression</code> for the <code>Binary Classification</code> of emails as ham (non-spam) (0) or spam (1).

<h2>PROBLEM STATEMENT<h2>

>In today's world, email (electronic mail) is one of the most prevalent forms of digitial. Almost all online sites and portals use emails to communicate with and notify their users, and mainly to announce any promotions or extend offers. For the purpose of advertising, several unsolicited, junk mails are sent in bulk, known as 'spam mails'. Spam mails clutter inboxes and can also pose huge risks as they may contain malicious attachments or links which download malware (viruses, ransomware, spyware, etc.) on the recipient's device on  being clicked. Such mails are used for identity nd data theft, which can be used to cause financial losses and severe damage to reputation. It is therefore extremely crucial for users and systems to detect/recognise spam mails and filter/delete them. 

<h2>PROPOSED SOLUTION<h2>

>In order to detect and filter out spam mails, the device/system must have robust built-in spam filters that are properly enabled to instantaneously detect spam mails and automatically shift them to a separate folder so as to safeguard the user's privacy and device. Such a filter can be achieved by developing a machine learning model and rigorously training it to differentiate between spam and non-spam mails. The machine learning model will then be able to effectively detect spam mails with high accuracy, thereby protecting the user and keeping the user alert and aware.

<h3>BALANCING THE DATASETS<h3>

>When training a machine learning model, it is extremely important to use datasets that are balanced or uniformly distributed. This is because using an unbalanced dataset can cause the model to become biased towards the majority class, and due to such influence it may predict the minority class poorly, leading to highly inaccurate predictions, analyses, and results. But if the datasets are balanced, the model will show predictions with high accuracy, fairness, and without bias. This ensures fair representation of all the classes involved, heightening the model's performance. 

```python
h1=ec1[ec1['Category']=='ham']
s1=ec1[ec1['Category']=='spam']
bh1=h1.sample(n=len(s1),random_state=5)
ec1=pa.concat([bh1,s1]).reset_index(drop=True)
mails1=ec1.where((pa.notnull(ec1)), '')
h2=ec2[ec2['spam']==0]
s2=ec2[ec2['spam']==1]
bh2=h2.sample(n=len(s2),random_state=5)
ec2=pa.concat([bh2,s2]).reset_index(drop=True)
mails2=ec2.where((pa.notnull(ec2)), '')
h3=ec3[ec3['label']==0]
s3=ec3[ec3['label']==1]
bh3=h3.sample(n=len(s3),random_state=5)
ec3=pa.concat([bh3,s3]).reset_index(drop=True)
mails3=ec3.where((pa.notnull(ec3)), '')
```

<h3>FEATURE EXTRACTION<h3>

>- Performing feature extraction and feature engineering: this will help the model to better learn the parameters of the training datasets, ensuring better training and consistent scaling. 
>- Any stopwords that do not have any significant meaning will be ignored by the model for better performance. In this way, there is no need to perform further data cleaning to remove stopwords which thus prevents information loss.

```python
fe=TV(min_df=1,stop_words='english',lowercase=True)
MSG1trf=fe.fit_transform(MSG1tr)
MSG1tf=fe.transform(MSG1t)
CTG1tr=CTG1tr.astype('int')
CTG1t=CTG1t.astype('int')
MSG2trf=fe.fit_transform(MSG2tr)
MSG2tf=fe.transform(MSG2t)
CTG2tr=CTG2tr.astype('int')
CTG2t=CTG2t.astype('int')
MSG3trf=fe.fit_transform(MSG3tr)
MSG3tf=fe.transform(MSG3t)
CTG3tr=CTG3tr.astype('int')
CTG3t=CTG3t.astype('int')
TMSG1trf=fe.fit_transform(TMSG1tr)
TMSG1tf=fe.transform(TMSG1t)
TCTG1tr=TCTG1tr.astype('int')
TCTG1t=TCTG1t.astype('int')
```
